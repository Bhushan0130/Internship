```python
url_3 = 'https://www.google.co.in/imghp?hl=en&authuser=0&ogbl'
driver_3 = webdriver.Chrome("D:/Bhushan Sharma/Internship/chromedriver_win32/chromedriver.exe")
driver_3.get(url_3)
driver_3.execute_script("window.scrollTo(0, 0);")
driver_3.maximize_window()
```

```python
u_agent = u_agnt = {
'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36',
'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
'Accept-Charset': 'ISO-8859-1, utf-8; q=0.7, *; q=0.3',
'Accept-Encoding':'none',
'Accept-Language': 'en-US, en;q=0.8',
'Connection': 'keep-alive' }
```

```python
image_folder = 'Google_Image'

def main():
    if not os.path.exists(image_folder):
        os.mkdir(image_folder)
    download_images()

def download_images():
    data = input('Enter a keyword to search: ')
    num_images = int(input('Enter the number of images you want: '))
    search_url = url_3 + 'q=' + data
    
    response = requests.get(search_url, headers = u_agnt)
    html = response.text
    
    
    b_soup = BeautifulSoup(html, 'html.parser')
    results = b_soup.findAll('img', {'class' : 'rg_i Q4LuWd'})
    
    count = 0
    image_links = []
    for res in results:
        try:
            link = res['data-src']
            image_links.append(link)
            count = count +1
            if (count >= num_images):
                break
        except KeyError:
            continue
            
        
    
    for i, link in enumerate(image_links):
        response = requests.get(link)
        
        imagename = image_folder + '/' + data + str(i+1)+'.jpg'
        with open(imagename, 'wb') as file:
            file.write(response.content)
    
    print('Download Completed')
    
if __name__ == '__main__' :
    main()
```

```python
url_list_10 = ['https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-12-13&to=2021-12-16&guests=2&page=1' ,
                'https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-12-13&to=2021-12-16&guests=2&page=2' ,
                'https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-12-13&to=2021-12-16&guests=2&page=3' ,
                'https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-12-13&to=2021-12-16&guests=2&page=4' ]
```

```python
hostal_name = []
distance = []
rating_10 = []
reviews_10 = []
total_review = []
facility = []
# private = []
for u in url_list_10:
    driver_10.get(u)
    time.sleep(4)

    name_tag =  driver_10.find_elements_by_xpath("//h2[@class= 'title title-6']/a")
    distance_tag =  driver_10.find_elements_by_xpath("//a[@class='show-on-map']/span[@class ='description']")
    rating_tag = driver_10.find_elements_by_xpath("//div[@class = 'rating rating-summary-container big']/div[@class = 'score orange big']")
    review_tag = driver_10.find_elements_by_xpath("//div[@class = 'keyword']")
    count_review_tag = driver_10.find_elements_by_xpath("//div[@class  = 'reviews']")
    fac_tag = driver_10.find_elements_by_xpath("//div[@class = 'rating-factors prop-card-tablet rating-factors small']")

    for i, j, rate, n_review, fac in zip(name_tag, distance_tag, rating_tag, count_review_tag, fac_tag ):
        hostal_name.append(i.text)
        distance.append(j.text)
        rating_10.append(rate.text)
        total_review.append(n_review.text.split(' ')[0])
        facility.append(fac.text.split('\n'))
           
    for review in review_tag[3:]:
        reviews_10.append(review.text)
```
