### Question 1
Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos   You need to find following details:

```python
# import all required libraries
import selenium 
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
import time

from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url_1= "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver_1 = webdriver.Chrome("D:/Bhushan Sharma/Internship/chromedriver_win32/chromedriver.exe")
driver_1.get(url_1)
driver_1.maximize_window()

delay = 10
WebDriverWait(driver_1, delay).until(EC.presence_of_element_located((By.CLASS_NAME, 'reference')))
# Found multiple elements
tag1 = driver_1.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']/tbody/tr/td")
```

```python
rank = []
name = []
artist = []
view=[]
uploaded = []

for i in tag1[0:180:6]:
    rank.append(int(i.text.replace('.', '')))

for n in tag1[1:181:6]:
    name.append(n.text)

for v in tag1[2:182:6]:
    artist.append(v.text)

for v in tag1[3:183:6]:
    view.append(v.text)

for u in tag1[4:184:6]:
    uploaded.append(u.text)
    
print(len(rank))
print(len(name))
print(len(artist))
print(len(view))
print(len(uploaded))
```

```python
df1 = pd.DataFrame()  # DataFrame for extracted elements

# Imputation of data into dataframe
df1['Rank'] = rank
df1['Name'] = name
df1['Artist'] = artist
df1['View'] = view
df1['uploaded'] = uploaded
df1
```
