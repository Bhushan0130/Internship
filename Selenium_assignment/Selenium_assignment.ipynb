```python
# lets first connect to the webdriver
driver = webdriver.Chrome("E:/chromedriver.exe")  # Downloaded Chrome path
url = "https://www.naukri.com/"
driver.get(url)
driver.maximize_window()

search_job = driver.find_element_by_id('qsb-keyword-sugg')
time.sleep(2)
search_job.send_keys("Data Analyst")   # Enter “Data Analyst” in “Skill, Designations, Companies” field
search_loc = driver.find_element_by_xpath("//input[@id = 'qsb-location-sugg']")

time.sleep(2)
search_loc.send_keys("Bangalore")    # Enter “Bangalore” in “enter the location” field

search_btn = driver.find_element_by_xpath("//div[@class='search-btn']/button")
time.sleep(2)    
search_btn.click()     # click the search button
```

```python
time.sleep(5) 
cross_chatbot = driver.find_element_by_xpath("//div[@class = 'crossIcon chatBot chatBot-ic-cross']")
cross_chatbot.click()   # Crossing the chatbot

time.sleep(5)
location_cross = driver.find_element_by_xpath("//span[@id = 'block']")
location_cross.click()   # Crossing the location to the "later" option 

job_titles = []
company_names = []
experience_list = []
locations_list = []

# extraction of  all the tags
titles_tags = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")  
company_n_tags = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
exp_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
location_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]") 

# Appending List for their respective list 
for i in range (10):
    job_titles.append(titles_tags[i].text)
    experience_list.append(exp_tags[i].text)
    company_names.append(company_n_tags[i].text)
    locations_list.append(location_tags[i].text) 

# print(job_titles)
# print('___________________________________________________________________________________________')
# print(company_names)
# print('___________________________________________________________________________________________')
# print(experience_list)
# print('___________________________________________________________________________________________')
# print(locations_list)
# print('___________________________________________________________________________________________')

# -------------------------------------------------Preprating DataSet --------------------------------------------------
df = pd.DataFrame()
df['Job Title'] = job_titles
df['Company Name'] = company_names
df['Experience'] = experience_list
df['Location'] = locations_list
df.set_index( [[*range(1,11)]] , inplace = True)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
print("____________________________________________ Details of 10 Records _____________________________________________")
df
```

```python
driver_8 = webdriver.Chrome("E:/chromedriver.exe")
url_8 = "https://www.amazon.in/"
driver_8.get(url_8)
driver_8.maximize_window()

time.sleep(4)
srch_input_8 = driver_8.find_element_by_xpath("//input[@id = 'twotabsearchtextbox']")
srch_input_8.send_keys('Laptop')
time.sleep(5)

search_btn_8 = driver_8.find_element_by_xpath("//input[@id = 'nav-search-submit-button']")
search_btn_8.click()

time.sleep(10)
Intel_Core_i7 = driver_8.find_element_by_xpath("//span[text()='Intel Core i7']")
Intel_Core_i7.click()                   # intel core i7 filter applying

time.sleep(5)
Intel_Core_i9 = driver_8.find_element_by_xpath("//span[text()= 'Intel Core i9']")
Intel_Core_i9.click()                    # intel core i8 filter applying

product_name_8 = driver_8.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
product_prize_8 = driver_8.find_elements_by_xpath("//span[@class = 'a-price-whole']")

laptop_name_8=[]
prize_8 = []
for i in range(10):
    try:
        laptop_name_8.append(product_name_8[i].text)
        prize_8.append(product_prize_8[i].text.replace(',',  ''))
    except:
        laptop_name_8.append('None')
        
# fetching all the product url
url_tag_8 = driver_8.find_elements_by_xpath("//a[@class = 'a-link-normal a-text-normal']")

laptop_url_8 = []

for url_8 in url_tag_8[0:10]:
    laptop_url_8.append(url_8.get_attribute('href'))   

laptop_rating_8 = []

for laptop_8 in laptop_url_8:       #iterating over each and every product url present in list laptop_url
    
    driver_8.get(laptop_8)
    time.sleep(2.5)
    
    try :                     # here we are fetching ratings using try and exception 
        rating_8 = driver_8.find_element_by_xpath("//span[@class = 'a-size-medium a-color-base']")
                                                                
        laptop_rating_8.append(rating_8.text)
    except:
        laptop_rating_8.append('-')    #if rating is not present then append '-'
```
