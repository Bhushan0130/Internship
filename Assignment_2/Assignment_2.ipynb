## Answer_1

#### Python program to display all the header tags from‘en.wikipedia.org/wiki/Main_Page’

```python
# load Libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url_1 = 'https://en.wikipedia.org/wiki/Main_Page'
r_1 = requests.get(url_1)
r_1   # Here we are geting Response [200],  code indicates that the request has succeeded
```

```python
html_content_1 = r_1.content
soup_1 = BeautifulSoup(html_content_1, 'html.parser')
# print(soup_1.prettify())
```

```python
# HTML defines six levels of headings h1, h2, h3, h4, h5, h6 
header_type = ['h1', 'h2', 'h3', 'h4', 'h5' ,'h6']
header_name = soup_1.find_all(header_type)

for header in header_name:
    print(header.name , ' : ', header.text.strip().replace('expanded','').replace('collapsed', '').strip())
```

## Answer_2

#### Python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame.

```python
url_2 = 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'
r_2 = requests.get(url_2)
r_2
```

```python
html_content_2 = r_2.content
soup_2 = BeautifulSoup(html_content_2, 'html.parser')
# print(soup_2.prettify())
```

```python
scrap_data = soup_2.find_all('div', class_ = 'lister-item mode-advanced')
movie_name = [] 
year_of_release = []
IMDB_rating = []


for i in range(100) :
    title_2 = scrap_data[i].find('h3', class_ = 'lister-item-header')
    movie_2 = title_2.find('a').text.strip()
    movie_name.append(movie_2)
    
    y_release_2 = scrap_data[i].find('h3', class_ = 'lister-item-header')
    year_2 = y_release_2.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '').replace('I ', '')  
    year_of_release.append(int(year_2))
    
    IMDB_2 = scrap_data[i].find('div', class_ = 'inline-block ratings-imdb-rating')
    rating_2 = IMDB_2.find('strong').text
    IMDB_rating.append(round(float(rating_2), 1 ))

# print(len(movie_name))
# print(len(year_of_release))
# print(len(IMDB_rating))
```

```python
# print('--------------------------------------------Movie Name -------------------------------------------')
# print(movie_name)
# print('__________________________________________________________________________________________________')
# print('------------------------------------------year of release ----------------------------------------')
# print(year_of_release)
# print('__________________________________________________________________________________________________')
# print('------------------------------------------IMDB rating --------------------------------------------')
# print(IMDB_rating)
# print('__________________________________________________________________________________________________')
```

```python
head_2 = soup_2.find_all(  'h1',   class_ = 'header')
headline_2 = head_2[0].text.replace('\n',' ').replace ('1000', '100')
```

```python
df_IMDB_100 = pd.DataFrame( data = {'Movie Name' : movie_name, 'IMDB Rating': IMDB_rating,
                                    'Year of release' : year_of_release}) 

df_IMDB_100.set_index( [ [*range(1,101)]] , inplace = True )

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
print('________________________________', headline_2,'____________________________________')
df_IMDB_100
```

## Answer_3

#### Python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and make data frame

```python
url_3 = 'https://www.imdb.com/india/top-rated-indian-movies/'
r_3 = requests.get(url_3)
r_3
```

```python
html_content_3 = r_3.content
soup_3 = BeautifulSoup(html_content_3, 'html.parser')
# print(soup_3.prettify())
```

```python
movie_scrap_3 = soup_3.find_all('td', class_ = 'titleColumn')
rating_3= soup_3.find_all('td', class_ = 'ratingColumn imdbRating')

movie_name_3 = []
year_of_release_3 = []
IMDB_rating_3 = []

for i in range (100):
    title_3 = movie_scrap_3[i].find('a').text.strip()
    movie_name_3.append(title_3)
    
    release = movie_scrap_3[i].find('span', class_ = 'secondaryInfo').text.replace('(','').replace(')', '')
    year_of_release_3.append(int(release))
    
    IMDB_3 = rating_3[i].text.strip()
    IMDB_rating_3.append(float(IMDB_3))

# print(len(movie_name_3))
# print(len(year_of_release_3))
# print(len(IMDB_rating_3))

title_3= soup_3.find_all('h1', class_ = 'header')
headline_3 = 'IMDb ' + title_3[0].text
```

```python
df_IMDB_100_Indian = pd.DataFrame()
df_IMDB_100_Indian['Movie Name'] = movie_name_3
df_IMDB_100_Indian['IMDB Rating'] = IMDB_rating_3
df_IMDB_100_Indian['Year of release'] = year_of_release_3
df_IMDB_100_Indian.set_index( [ [*range(1,101)]] , inplace = True )
```
