```python
# Operation for Sarees

urls = ['https://www.amazon.in/s?k=Sarees+%28women%29&ref=nb_sb_noss', 
       'https://www.amazon.in/s?k=Sarees+%28women%29&page=2&qid=1648723265&ref=sr_pg_1',
       'https://www.amazon.in/s?k=Sarees+%28women%29&page=3&qid=1648723446&ref=sr_pg_3',
       'https://www.amazon.in/s?k=Sarees+%28women%29&page=4&qid=1648723478&ref=sr_pg_4',
       'https://www.amazon.in/s?k=Sarees+%28women%29&page=5&qid=1648723494&ref=sr_pg_5']

product_url = []
image = []
```

```python
def product_url_func (urls):    
    for url in urls:
        driver.get(url)

        urls_link = driver.find_elements_by_xpath("//a[@class = 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
        for i in urls_link:
            product_url.append(i.get_attribute('href'))
```

```python
def image_link(prod_url):
    driver.get(prod_url)
    image_tag = driver.find_element_by_xpath("//div[@id='imgTagWrapperId']/img")
    image.append(image_tag.get_attribute('src'))
```

```python
def create_folder(dir):
    current_path=os.getcwd()
    new=os.path.join(current_path,dir)
    if not os.path.exists(new):
        os.makedirs(new)
        
# create_folder('Sarees_Images')
# create_folder('Trousers_Images')
# create_folder('Jeans_Images')  
```

```python
def download_image(image_list, folder_name, string1):
    # Dowloading images
    for index, link in enumerate(image_list):
        print('Downloading {0} '.format(index+1))
        try:
            response=requests.get(link)
        except: pass
        
        with open('{}/{}_{}.jpeg'.format(folder_name,string1, index+1),"wb") as file:
            file.write(response.content)
```

```python
product_url_func(urls)
# print(product_url[0])
print(len(product_url))
```

```python
for i in product_url:
    try:
        image_link(i)
    except: pass
# print(image[0])
print(len(image))
```

```python
folder1 = 'Scraped_images'
# folder2 = 'Trousers_Images'
# folder3 = 'Jeans_Images'

create_folder('Scraped_images')
# create_folder('Trousers_Images')
# create_folder('Jeans_Images') 
```
